```python
number = '15'
version = 'separate' # algo -> 'same_time' or 'separate'
extension = '' # extension de nom de fichier -> '_s_t' or ''

Occupancy_perf = np.load(f"Results_{version}/condition_{number}/Occupancy_perf_condition_{number}{extension}.npy")
Abandonment_perf = np.load(f"Results_{version}/condition_{number}/Abandonment_perf_condition_{number}{extension}.npy")
Waiting_time_perf = np.load(f"Results_{version}/condition_{number}/Waiting_time_perf_condition_{number}{extension}.npy")
Detour_time_perf = np.load(f"Results_{version}/condition_{number}/Detour_time_perf_condition_{number}{extension}.npy")
history_sharing = np.load(f"Results_{version}/condition_{number}/history_sharing_condition_{number}{extension}.npy")

Time = [i for i in range(len(Occupancy_perf[0]))]

# print("occupancy performance : ",Occupancy_perf.shape,type(Occupancy_perf),"\n",Occupancy_perf, "\n")
# print("abandonment performance : ",Abandonment_perf.shape,type(Abandonment_perf),"\n",Abandonment_perf,"\n")
# print("waiting time performance : ",Waiting_time_perf.shape,type(Waiting_time_perf),"\n",Waiting_time_perf,"\n")
# print("detour performance : ",Detour_time_perf.shape,type(Detour_time_perf),"\n",Detour_time_perf, "\n")
# print("history sharing : ",history_sharing.shape,type(history_sharing),"\n",history_sharing, "\n")
```

## Graphs

Run all the following cells to compute the different graphs. 

#### Occupancy

```python
plt.figure(1)

plt.plot(Time, Occupancy_perf[0])
plt.plot(Time, Occupancy_perf[1])
plt.plot(Time, Occupancy_perf[2])
plt.title('Occupancy during simulation')
plt.legend(['Occupancy of vehicles = 0', 'Occupancy of vehicles = 1', 'Occupancy of vehicles = 2'])
plt.xlabel('Time step')
plt.ylabel('Number of vehicules')
plt.show()
```

#### Abandonment

```python
plt.figure(2)

plt.plot(Time, Abandonment_perf)
plt.title('Abandonment')
plt.xlabel('Time step')
plt.ylabel('Abandonment')
plt.show()

print(f'abandonment {int(Abandonment_perf[-1])} requests')
```

#### Waiting Time

```python
plt.figure(3)

Request_w = [i for i in range(len(Waiting_time_perf))]
mean_w = np.ones([len(Waiting_time_perf)])*np.mean(Waiting_time_perf)*60
plt.scatter(Request_w, Waiting_time_perf*60,s=0.2)
plt.plot(Request_w, mean_w, color='orange')
plt.title('Waiting Time [min]')
plt.xlabel('Requests')
plt.ylabel('Waiting time [min]')
plt.legend(['waiting times','mean'])
plt.show()

print(f'mean waiting time : {mean_w[0]:.4f}[min]')
print(len(Waiting_time_perf))
```

#### Detour Time

```python
if len(Detour_time_perf) == 0:
    print('No sharing trips, so no detour time saved.')
    
else :
    plt.figure(4)

    Request_d = [i for i in range(len(Detour_time_perf))]
    mean_d = np.ones([len(Detour_time_perf)])*np.mean(Detour_time_perf)*60
    plt.scatter(Request_d, Detour_time_perf*60,s=1)
    plt.plot(Request_d, mean_d, color='orange')
    plt.title('Detour Time')
    plt.xlabel('Requests')
    plt.ylabel('Detour time [min]')
    plt.show()

    print(f'mean detour time : {mean_d[0]:.4f}[min]')
```

```python
number_exeed = []


for i in range(len(Detour_time_perf)):
    if Detour_time_perf[i]*60>20:
        number_exeed.append(i)
        
print(len(number_exeed))
```
